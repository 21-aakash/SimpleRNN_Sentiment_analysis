```python
from tensorflow.keras.preprocessing.text import one_hot # type: ignore
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### one hot representation 

one_hot_repr= [one_hot(words, voc_size)for words in sent]
one_hot_repr

## indexs
```

```python
from tensorflow.keras.layers import Embedding # type: ignore

from tensorflow.keras.utils import pad_sequences # type: ignore
from tensorflow.keras.models import Sequential # type: ignore
```

```python
import numpy as np
```

```python
## making all sentence of same length 8 , as max is 8 and addding 0 as padding 

sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation

## dimensions 300 or 10 for ex

dim=10
```

```python
## making the model 

model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')

# optimizer adam , loss fun mse
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
